```python
#Вариант 10: Flights (авиаперевозки)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Настройки
sns.set_style('whitegrid')
%matplotlib inline
```

```python
BOLD = '\033[1m'
RESET = '\033[0m'

#ОБНАРУЖЕНИЕ
print("="*30)
print(f"{BOLD}ЭТАП 1: Discover (Обнаружение){RESET}")
print("="*30)

# Загрузка данных
df = sns.load_dataset('flights')

# 1.1 Размер и первые строки
print(f"\n {BOLD}Размер исходных данных:{RESET}", df.shape)
print(f"\n {BOLD}Первые 5 строк:{RESET}")
display(df.head())

print(f"\n {BOLD}Последние 5 строк:{RESET}")
display(df.tail())

# 1.2 Общая информация
print(f"\n {BOLD}Информация о датасете:{RESET}")
df.info()

# 1.3 Статистическое описание
print(f"\n {BOLD}Статистическое описание:{RESET}")
display(df.describe())

# 1.4 Проверка пропусков
print(f"\n {BOLD}Проверка пропусков:{RESET}")
print(df.isnull().sum())

# 1.5 Проверка дубликатов
print(f"\n {BOLD}Проверка дубликатов:{RESET}")
print(f" {BOLD}Количество дубликатов:{RESET} {df.duplicated().sum()}")

# 1.6 Уникальные значения
print(f"\n {BOLD}Уникальные значения в year:{RESET}", sorted(df['year'].unique()))
print(f" {BOLD}Уникальные значения в month:{RESET}", df['month'].unique())
print(f" {BOLD}Диапазон passengers:{RESET} от", df['passengers'].min(), "до", df['passengers'].max())
```

```python
#СТРУКТУРИРОВАНИЕ

print(f"\n{BOLD}{'='*36}{RESET}")
print(f"{BOLD}ЭТАП 2: STRUCTURE (Структурирование){RESET}")
print(f"{BOLD}{'='*36}{RESET}")

df_struct = df.copy()

df_struct['date_str'] = df_struct['year'].astype(str) + ' ' + df_struct['month'].astype(str)
df_struct['date'] = pd.to_datetime(df_struct['date_str'], format='%Y %b')

print(f"\n{BOLD} Новый столбец date (первые 12 строк):{RESET}")
display(df_struct[['year', 'month', 'date']].head(12))

df_struct.set_index('date', inplace=True)
df_struct.sort_index(inplace=True)
```

```python
#ОЧИСТКА

print(f"\n{BOLD}{'='*23}{RESET}")
print(f"{BOLD}ЭТАП 3: CLEAN (Очистка){RESET}")
print(f"{BOLD}{'='*23}{RESET}")

df_clean = df_struct.copy()

print(f"{BOLD}Пропусков до заполнения:{RESET} {df_clean.isnull().sum().sum()}")
print(f"{BOLD}Дубликатов до удаления:{RESET} {df_clean.duplicated().sum()}")

df_clean.drop(columns=['date_str'], inplace=True, errors='ignore')

print(f"{BOLD}Размер после очистки:{RESET} {df_clean.shape}")
```

```python
#ОБОГАЩЕНИЕ

print(f"\n{BOLD}{'='*27}{RESET}")
print(f"{BOLD}ЭТАП 4: ENRICH (Обогащение){RESET}")
print(f"{BOLD}{'='*27}{RESET}")

df_enrich = df_clean.copy()

# Признак 1: номер месяца
df_enrich['month_num'] = df_enrich.index.month
print(f"{BOLD}Создан признак -{RESET}'month_num'")

# Признак 2: season
def get_season(month):
    if month in [12, 1, 2]:
        return 'winter'
    elif month in [3, 4, 5]:
        return 'spring'
    elif month in [6, 7, 8]:
        return 'summer'
    else:
        return 'autumn'

df_enrich['season'] = df_enrich['month_num'].apply(get_season)
print(f"{BOLD}Создан признак -{RESET}'season'{RESET}")

# Признак 3: лаг 1 месяц (shift)
df_enrich['passengers_lag1'] = df_enrich['passengers'].shift(1)
print(f"{BOLD}Создан признак -{RESET}'passengers_lag1'{RESET}")

# Дополнительные признаки
df_enrich['passengers_lag12'] = df_enrich['passengers'].shift(12)
print(f"{BOLD}Создан признак -{RESET}'passengers_lag12'")

print(f"\n{BOLD}Обогащённые данные (первые 15 строк):{RESET}")
display(df_enrich.head(15))
```

```python
#ВАЛИДАЦИЯ

print(f"\n{BOLD}{'='*28}{RESET}")
print(f"{BOLD}ЭТАП 5: VALIDATE (Валидация){RESET}")
print(f"{BOLD}{'='*28}{RESET}")

df_final = df_enrich.reset_index()

print(f"\n{BOLD}Пропуски в финальном датасете:{RESET}")
print(df_final.isnull().sum())

print(f"\n{BOLD}Размер финального датасета:{RESET} {df_final.shape}")

output_filename = 'cleaned_flights.csv'
df_final.to_csv(output_filename, index=False)
print(f"\n{BOLD}Датасет сохранён: '{output_filename}'{RESET}")
print(f"\n{BOLD}ПОЛНЫЙ ЦИКЛ EDA ЗАВЕРШЁН{RESET}")
```
