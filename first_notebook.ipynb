```python
#Задание 3.4: Операции с числами
# Арифметические операции
a = 15
b = 4
print("a =", a, ", b =", b)
print("Сложение:", a + b)
print("Вычитание:", a - b)
print("Умножение:", a * b)
print("Обычное деление:", a / b) # Всегда дает float
print("Целочисленное деление:", a // b) # Отбрасывает дробную часть
print("Остаток от деления:", a % b) # Полезно для проверки четности
print("Возведение в степень:", a ** 2) # Квадрат числа
print("Квадратный корень:", a ** 0.5)
# Операции с присваиванием
счетчик = 10
счетчик += 5 # То же, что счетчик = счетчик + 5
print("Счетчик после +=5:", счетчик)
счетчик *= 2 # Умножаем на 2
print("Счетчик после *=2:", счетчик)
```

```python
#Задание 3.5: Работа со строками
# Создаем строки
приветствие = "Hello"
имя = "World"
# Конкатенация (сложение строк)
полное_приветствие = приветствие + " " + имя + "!"
print(полное_приветствие)
# Умножение строк
разделитель = "-" * 30
print(разделитель)
# Методы строк
текст = " Python для анализа данных "
print("Исходный текст: '", текст, "'", sep="")
print("В верхнем регистре:", текст.upper())
print("В нижнем регистре:", текст.lower())
print("С заглавной буквы:", текст.strip().capitalize())
print("Без пробелов по краям: '", текст.strip(), "'", sep="")
print("Замена 'данных' на 'информации':",
текст.replace("данных", "информации"))
# F-строки (современный способ форматирования)
язык = "Python"
версия = 3.9
print(f"Я изучаю {язык} версии {версия}")
print(f"Площадь круга радиуса 5: {3.14159 * 5 ** 2:.2f}") #:.2f - округление до 2 знаков
```

```python
#Задание 3.6: Ввод данных от пользователя
# Функция input() всегда возвращает строку
имя_пользователя = input("Как вас зовут? ")
print(f"Привет, {имя_пользователя}!")
# Для чисел нужно преобразовывать тип
возраст_строка = input("Сколько вам лет? ")
возраст_число = int(возраст_строка) # Преобразуем строку в целое число
год_рождения = 2024 - возраст_число
print(f"Вы родились примерно в {год_рождения} году")
# Можно сразу преобразовывать
рост = float(input("Ваш рост в метрах (например, 1.75): "))
вес = float(input("Ваш вес в кг: "))
имт = вес / (рост ** 2)
print(f"Ваш индекс массы тела: {имт:.1f}")
```

```python
#Задание 3.7: Мини-проект "Калькулятор ИМТ"
#Создайте новую ячейку и напишите полноценную программу:
# ============================================
# КАЛЬКУЛЯТОР ИНДЕКСА МАССЫ ТЕЛА (ИМТ)
# ============================================
print("=" * 40)
print("КАЛЬКУЛЯТОР ИНДЕКСА МАССЫ ТЕЛА")
print("=" * 40)
# 1. Получаем данные
имя = input("Введите ваше имя: ")
рост = float(input("Введите ваш рост в метрах (например, 1.75):"))
вес = float(input("Введите ваш вес в кг: "))
# 2. Рассчитываем ИМТ
имт = вес / (рост ** 2)
# 3. Определяем категорию
if имт < 18.5:
 категория = "Недостаточный вес"
elif имт < 25:
 категория = "Нормальный вес"
elif имт < 30:
 категория = "Избыточный вес"
else:
 категория = "Ожирение"
# 4. Выводим результат
print("\n" + "=" * 40)
print("РЕЗУЛЬТАТЫ:")
print("=" * 40)
print(f"Имя: {имя}")
print(f"Рост: {рост} м")
print(f"Вес: {вес} кг")
print(f"ИМТ: {имт:.1f}")
print(f"Категория: {категория}")
print("=" * 40)
# 5. Рекомендации
print("\nСправка по категориям ИМТ:")
print("• Менее 18.5: Недостаточный вес")
print("• 18.5-24.9: Нормальный вес")
print("• 25-29.9: Избыточный вес")
print("• 30 и более: Ожирение")
```
